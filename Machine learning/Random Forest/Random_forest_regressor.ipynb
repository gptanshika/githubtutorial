```python
num_feature = [feature for feature in X.columns if X[feature].dtype !='O']
print(num_feature)
```

```python
cat_features = [feature for feature in X.columns if X[feature].dtype =='O']
print(cat_features)
```

```python
des_features = [feature for feature in num_feature if len(X[feature].unique())<=25]
print(des_features)
```

```python
con_feature = [feature for feature in num_feature if feature not in des_features]
print(con_feature)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder

labelencoder = LabelEncoder()

X_train['model'] = labelencoder.fit_transform(X_train['model'])
X_test['model'] = labelencoder.transform(X_test['model'])
```

```python
X_train.head(5)
```

```python
from sklearn.compose import ColumnTransformer

column_trans = ColumnTransformer([
    ('Classification',OneHotEncoder(drop='first'),cat_features),
    ('Numerical',StandardScaler(),con_feature)
],remainder='passthrough')
```

```python
X_train = column_trans.fit_transform(X_train)
X_test = column_trans.transform(X_test)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)

print(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
def eval_model(true,predicted):
    mae = mean_absolute_error(true,predicted)
    mse = mean_squared_error(true,predicted)
    r2_score_2 = r2_score(true,predicted)
    return mae,mse,r2_score_2
```

```python
models = {
    "Linear Regression" : LinearRegression(),
    "Ridge" : Ridge(),
    "Lasso" : Lasso(),
    "K-Neighbors Regressor" : KNeighborsRegressor(),
    "Decision Tree Regressor" : DecisionTreeRegressor(),
    "Random Forest Regressor" : RandomForestRegressor()
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train,y_train)

    y_pred = model.predict(X_test)
    print(f'{i} = ',eval_model(y_test,y_pred))
```
